% Reguła do sprawdzenia połączenia bezpośredniego
bezposrednie_polaczenie(X, Y) :-
    polaczenie(X, Y, _).

% Reguła do obliczenia kosztu podróży
koszt_podrozy(X, Y, C) :-
    polaczenie(X, Y, C).

% Sprawdzenie połączenia wychodzącego z X do Y
połączenie_wychodzące(X, Y, C) :-
    polaczenie(X, Y, C).

% Rekurencyjna reguła dla szukania trasy
trasa(X, Y, [X, Y], C) :-
    połączenie_wychodzące(X, Y, C).

trasa(X, Y, [X | T], C) :-
    połączenie_wychodzące(X, Z, C1),
    Z \= Y, % Z nie może być Y (nie chcemy, żeby zatrzymało się na Y)
    trasa(Z, Y, T, C2),
    C is C1 + C2.

% Znalezienie najkrótszej trasy
najkrotsza_trasa(X, Y, Trasa, Koszt) :-
    setof((C, T), trasa(X, Y, T, C), Wyniki),
    Wyniki = [(Koszt, Trasa) | _].

% Reguła sprawdzająca, czy produkt jest dostępny
dostepny_produkt(X) :-
    produkt(X, Ilosc, _),
    Ilosc > 0.

% Reguła przenoszenia produktu do innego sektora
przenies_produkt(X, NowySektor) :-
    retract(produkt(X, Ilosc, _)), % Usunięcie starego faktu
    assertz(produkt(X, Ilosc, NowySektor)). % Dodanie nowego faktu z nowym sektorem

% Reguła uzupełniania stanu produktu
uzupelnij_stan(X, N) :-
    produkt(X, Ilosc, Sektor),
    NowaIlosc is Ilosc + N,
    retract(produkt(X, Ilosc, Sektor)), % Usunięcie starego faktu
    assertz(produkt(X, NowaIlosc, Sektor)). % Dodanie nowego faktu z nową ilością

% Reguła do rekomendowania książek użytkownikowi na podstawie jego ocen i preferencji
polecane_ksiazki(X, Gatunek, ListaKsiążek) :-
    preferencje(X, Gatunek),                      % Sprawdzamy, czy użytkownik preferuje dany gatunek
    findall(Ksiazka, (ksiazka(Ksiazka, _, Gatunek), ocena(X, Ksiazka, Ocena), Ocena >= 4), ListaKsiążek).

% Dodanie nowej oceny
dodaj_ocene(Uzytkownik, Ksiazka, Ocena) :-
    assertz(ocena(Uzytkownik, Ksiazka, Ocena)).

% Usunięcie oceny
usun_ocene(Uzytkownik, Ksiazka) :-
    retract(ocena(Uzytkownik, Ksiazka, _)).

% Zmiana oceny
zmien_ocene(Uzytkownik, Ksiazka, NowaOcena) :-
    usun_ocene(Uzytkownik, Ksiazka),         % Usuwamy starą ocenę
    dodaj_ocene(Uzytkownik, Ksiazka, NowaOcena). % Dodajemy nową ocenę
